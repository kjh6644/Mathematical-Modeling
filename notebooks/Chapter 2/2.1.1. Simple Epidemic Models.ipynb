#### **1. 핵심 키워드 정리**

- **Epidemic(감염병)**: 인구 출생이나 자연 사망 등을 무시할 수 있을 만큼 '매우 짧은 시간 척도' 내에서 갑작스럽게 발생하여 급속히 퍼진 뒤 사라지는 질병의 outbreak를 의미

- **Endemic(풍토병/ 지역 내재 변이)**: Epidemic과 대비되는 개념으로, 특정 지역 내에서 질병이 소멸하지 않고 지속적으로 일정한 수준을 유지하며 존재

- **Immunity(면역)**: 인플루엔자나 홍역처럼 한 번 감염되고 회복된 후 획득하게 되는 질병에 대한 저항성

- **Compartmental Model(구획 모델)**: 복잡한 인구 집단을 질병의 상태라는 몇 개의 '구획'으로 단순화하여 나누고, 이 구획 간 시간적 이동 비율을 미분방정식으로 추적하는 수학적 접근

#### **2. 주요 내용**

1. **감염병의 역사적 파급력과 인구 동태학적 영향**
    - 과거 안토니우스 역병(로마)나 흑사병(유럽)은 사회 시스템을 붕괴
    - 코르테스의 아즈텍 정복 역시 무기의 우위가 아닌, '천연두'에 대한 아즈텍인의 면역 부재가 원인
    - 18세기 전 세계적인 인구 급증 현상은 농업 기술 발달뿐 아니라 교통의 발달로 질병이 순환하여 사람들의 **집단 면역(Immunities)**가 형성된 것도 원인

2. **Epidemic의 미스터리와 공통된 패턴**
    - 1665년 런던 대역병을 비롯한 수많은 감염병에서 독특한 패턴 발견
    - 질병은 갑자기 나타나 기하급수적으로 확산
    - 하지만 **어느 시점이 되면 전체 인구를 모두 감염시키지 않고 스스로 소멸**, 이 현상을 설명하기 위해 **Kermack-McKendrick 구획 모델 개발**

3. **수리역학 모델링의 실용적 목표**
    - 해당 감염병이 얼마나 심각해질 것인가? (총 감염자 수 예측)
    - 의료 시스템이 감당해야 할 최대 환자 수는 동시에 몇 명인가? (Peak size 예측)
    - 이 유행은 언제 끝날 것인가? (Duration 예측)
    - 환자 격리나 통제 정책이 실제로 유행을 얼마나 억제할 수 있는가? (정책 효과 시뮬레이션)
