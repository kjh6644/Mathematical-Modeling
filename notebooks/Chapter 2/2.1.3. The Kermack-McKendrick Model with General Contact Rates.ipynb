#### **1. 핵심 키워드 정리**

- **접촉 함수(Contact Function), $C(N)$**: 평균적인 한 개인이 단위 시간당 접촉하여 질병을 전파할 수 있는 총 사람의 수($C'(N) \ge 0$)

- **전파율 파라미터(Transmission Rate Parameter), $\beta(N)$**: 인구 한 명당 단위 시간 접촉률로, $\beta(N) = C(N) / N$으로 정의 ($\beta'(N) \le 0$)

- **회복 비율(Recovery Fraction), $f$**: 감염자(I) 상태를 벗어나는 사람 중 질병으로 사망하지 않고 면역을 얻어 회복하는 사람의 비율

- **기본 재생산지수(Basic Reproduction Number), $\mathcal{R}_0$**: 완전히 감수성인 인구 집단(K)에 처음 한 명의 감염자가 들어왔을 때, 이 감염자가 회복 또는 사망하기 전까지 유발하는 2차 감염자의 평균 수

- **유효 재생산지수(Running Reproduction Number), $\mathcal{R}^*$**: 유행이 진행 중인 임의의 시간 $t$에서, 새롭게 감염된 한 사람이 유발하는 2차 감염자의 수로 감수성 인구 $S(t)$가 감소함에 따라 유동적으로 변함

#### **2. 주요 내용**

1. **$\beta(N)$ 도입**: 현실적인 접촉 패턴을 모델링하기 위해 도입. 특정 질병의 경우 총인구수와 무관하게 접촉 횟수가 일정한 '표준 발생률(Standard incidence, $C(N)= \lambda$)'를 따를 수도 있고, 어느 정도 인구가 커지면 접촉 횟수가 포화 상태가 되는 함수를 따를 수도 있음
    - 접촉 함수의 도입으로 필연적으로 총인구 $N$을 변수화해 야됨

2. **정성적 결론 도출**:
    - 유효 재생산지수 $\mathcal{R}^*$의 도함수를 분석하여, $\mathcal{R}^* > 1$일 때 유행 규모는 감소함을 증명
    - 미분방정식을 적분하여 최종적으로 질병으로 죽은 사람 수와 병에 걸렸던 사람 수 간의 선형 비례 관계를 찾아내며, 유행이 종식된 후에도 한 번도 병에 걸리지 않은 생존자($S_{\infty}>0$)가 존재함을 증명

#### **4. 수식 전개**

1. **접촉 함수의 특성과 조건**
접촉 함수 $C(N) = N \beta(N)$이며, 인구가 증가할수록 총 접촉수는 증가하거나 일정($C'(N) \ge 0$)하지만, 인구 밀도에 따른 포화 효과로 1인당 접촉률은 감소하거나 일정($\beta'(N) \le 0$)하다고 가정.
\begin{aligned}
C'(N) = \beta(N) + N \beta'(N) \ge 0
\end{aligned}

2. **일반화된 SIR 역학 모델**
$N = S +I + R$이며, $f=1$이면 단순 모델과 동일
\begin{aligned}
S' &= -\beta(N) S I \\
I' &= \beta(N) S I - \alpha I \\
N' &= - (1-f) \alpha i
\end{aligned}

3. **재생산지수**
초기 인구를 $K \approx S(0)$이라고 할 때, 기본 재생산지수는 다음과 같다.
\begin{aligned}
\mathcal{R}_0 = {K \beta(K) \over \alpha}
\end{aligned}
시간 $t$에서의 유효 재생산지수는 다음과 같다.
\begin{aligned}
\mathcal{R}^* = {S \beta(N) \over \alpha}
\end{aligned}
